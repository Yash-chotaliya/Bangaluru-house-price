```python
import pandas as pd
import numpy as np

df = pd.read_csv('./dataset/data.csv')
df
```

```python
df['area_type'].value_counts()
```

```python
df1 = df.drop(['area_type','availability','society','balcony'],axis=1)
df1
```

```python
df1.isnull().sum()
```

```python
# our dataset has 13320 rows so we can drop null rows (73,16,1)

df1.dropna(inplace=True)
df1
```

```python
df1['size'] = df1['size'].apply(lambda x : x.split(' ')[0])
df1.rename(columns={'size':'bhk'},inplace=True)
df1
```

```python
df1.total_sqft.unique()
```

here we have some range value in total_sqft column so we take the average of both 

```python
# this function checks if number is float or not

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df1[~df1.total_sqft.apply(is_float)]
```

```python
def convert_avg(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2 = df1.copy()
df2['total_sqft'] = df2['total_sqft'].apply(convert_avg)
```

```python
df2
```

```python
df2.info()
```

```python
df2['bhk'] = df2['bhk'].astype(int)
```

```python
# add price per sqft column

df3 = df2.copy()
df3['price_per_sqft'] = df3['price']*100000/df3['total_sqft']
df3.info()
```

```python
location_stats = df3['location'].value_counts()
location_stats
```

```python
location_stats_less_10 = location_stats[location_stats<10]
len(location_stats_less_10)
```

we have 1051 locations which has <10 rows

```python
# we have to replace location with other which have < 10 location counts

df3['location'] = df3['location'].apply(lambda x : 'other' if x in location_stats_less_10 else x)
```

```python
df3
```

```python
df4 = df3[~(df3['bath']>=df3['bhk']+2)]
df4 = df4[~((df4['total_sqft']/df4['bhk'])<300)]
```

```python
df4.price_per_sqft.describe()
```

```python
df4.total_sqft.describe()
```

```python
# remove outlier in price_per_sqft

def remove_pps_outlier(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        mean = np.mean(subdf.price_per_sqft)
        std = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf['price_per_sqft']>(mean-std)) & (subdf['price_per_sqft']<(mean+std))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df4 = remove_pps_outlier(df4)
df4['location'] = df4['location'].apply(lambda x:x.strip())
```

```python
df4
```

# model building

```python
dummies = pd.get_dummies(df4['location'])
dummies.head()
```

```python
df5 = pd.concat([df4,dummies.drop('Vittasandra',axis=1)],axis=1)
df5.drop(['location','price_per_sqft'],axis=1,inplace=True)
df5
```

```python
X = df5.drop('price',axis=1)
y = df5['price']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

shuffle = ShuffleSplit(n_splits=5, test_size=0.2)

cross_val_score(LinearRegression(), X, y, cv=shuffle)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model(X,y):
    algos={
        'linear_regression':{
            'model':LinearRegression(),
            'params':{
                
            }
        },
        'lasso':{
            'model':Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random','cyclic']
            }
        },
        'decision_tree':{
            'model':DecisionTreeRegressor(),
            'params':{
                'criterion':['mse','friedman_mse'],
                'splitter':['best','random']
            }
        }
    }
                                          
    scores=[]
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=10)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'],config['params'],cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
                        
    return pd.DataFrame(scores)
                                          
find_best_model(X,y)
```

from above 3 models we get best score in first model so we use first model for prediction

```python
def predict_price(bhk,total_sqft,bath,location):
    loc_index = np.where(X.columns==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = bhk
    x[1] = total_sqft
    x[2] = bath
    if loc_index>=0:
        x[loc_index] = 1
    
    return model.predict([x])[0]
```

```python
predict_price(2,1170.0,2.0,'Devarachikkanahalli')
```
