```python
import pandas as pd
import numpy as np

df = pd.read_csv('./dataset/data.csv')
df
```

```python
df['area_type'].value_counts()
```

```python
df1 = df.drop(['area_type','availability','society','balcony'],axis=1)
df1
```

```python
df1.isnull().sum()
```

```python
# our dataset has 13320 rows so we can drop null rows (73,16,1)

df1.dropna(inplace=True)
df1
```

```python
df1['size'] = df1['size'].apply(lambda x : x.split(' ')[0])
df1.rename(columns={'size':'bhk'},inplace=True)
df1
```

```python
df1.total_sqft.unique()
```

here we have some range value in total_sqft column so we take the average of both 

```python
# this function checks if number is float or not

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df1[~df1.total_sqft.apply(is_float)]
```

```python
def convert_avg(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2 = df1.copy()
df2['total_sqft'] = df2['total_sqft'].apply(convert_avg)
```

```python
df2
```

```python
df2.info()
```

```python
df2['bhk'] = df2['bhk'].astype(int)
```

```python
# add price per sqft column

df3 = df2.copy()
df3['price_per_sqft'] = df3['price']*100000/df3['total_sqft']
df3.info()
```

```python
location_stats = df3['location'].value_counts()
location_stats
```

```python
location_stats_less_10 = location_stats[location_stats<10]
len(location_stats_less_10)
```

we have 1051 locations which has <10 rows

```python
# we have to replace location with other which have < 10 location counts

df3['location'] = df3['location'].apply(lambda x : 'other' if x in location_stats_less_10 else x)
```

```python
df3
```

```python
df4 = df3[~(df3['bath']>df3['bhk']+2)]
df4 = df4[~((df4['total_sqft']/df4['bhk'])<300)]
```

```python
df4
```

```python
df4.total_sqft.describe()
```

```python
# remove outlier in total_sqft column

q1,q3 = df4.sort_values(by='total_sqft')['total_sqft'].quantile([0.25,0.75])
iqr = q3-q1

lower = q1-1.5*iqr
upper = q3+1.5*iqr

df4 = df4[(df4['total_sqft']>lower) & (df4['total_sqft']<upper)]
df4
```

```python
df4 = df4.sort_values(by='location')
df4
```

```python
df4.info()
```
